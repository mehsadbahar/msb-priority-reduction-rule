## 2. Basic Visualization

```python
import matplotlib.pyplot as plt
import numpy as np

def plot_sequence(vector):
    sequence = msb_priority_reduction(vector)
    
    plt.figure(figsize=(10, 6))
    
    for i, state in enumerate(sequence):
        plt.bar(range(len(vector)), state, alpha=0.5, label=f'Step {i}' if i < 3 else "")
    
    plt.xlabel('Position')
    plt.ylabel('Value')
    plt.title(f'Reduction Sequence for {vector}')
    plt.legend()
    plt.grid(True, alpha=0.3)
    plt.show()
    
    return sequence

# Test with different vectors
test_vectors = [[3, 2, 1, 1], [4, 0, 2], [2, 2, 2]]

for vec in test_vectors:
    seq = plot_sequence(vec)
    print(f"Vector {vec}: {len(seq)-1} steps\n")
```
